```python
def read_file(filename):
    tupe = []
    with open(filename, 'r') as f:
        content = f.readlines()
    for link in content:
        link = link.split('\n')[0]
        tupe.append(link)
    return tupe
```

```python
def ASCII(s):
    x = 0
    for i in xrange(len(s)):
        x += ord(s[i])*2**(8 * (len(s) - i - 1))
    return x
```

```python
def check(x, y):
    for i in x:
        if(i == y):
            return 0
    return 1
```

```python
def train_validate_test_split(df, train_percent=.6, validate_percent=.2, seed=None):
    np.random.seed(seed)
    perm = np.random.permutation(df.index)
    m = len(df)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.ix[perm[:train_end]]
    validate = df.ix[perm[train_end:validate_end]]
    test = df.ix[perm[validate_end:]]
    return train, validate, test
```

```python
def fill_Coord(df, train):
    df_tmp = df[df.coordinates == "[0.0, 0.0]"]
    index = df_tmp.index.get_values()
    for i in index:
        if(df.loc[i]['coordinates'] == "[0.0, 0.0]"):
            for j in range(len(train)):
                if(train.loc[j]['coordinates'] != "[0.0, 0.0]"):
                    if(train.loc[j]['district'] == df.loc[i]['district']):
                        if(train.loc[j]['ward'] == df.loc[i]['ward']):
                            tmp = train.loc[j]['coordinates']
                            df.loc[i,'coordinates'] = tmp
                            break
    return df
```

```python
def split_set(df):
    train, validate, test = train_validate_test_split(df)
    train = train.reset_index(drop=True)
    validate = validate.reset_index(drop=True)
    test = test.reset_index(drop=True)
    train.to_csv('train.csv')
    validate.to_csv('validate.csv')
    test.to_csv('test.csv')
```

```python
def Evir2HotKey(df):
    for i in range(len(df)):
        if(pd.isnull(df.loc[i]['environment'])):
            continue
        else:
            tmp = []
            evir_list = df.loc[i]['environment'].replace(' ,', "").split(',')
            for k in range(len(evir_ASCII)):
                num = None
                for evir in evir_list:
                    if(evir != ''):
                        if(long(ASCII(evir)) == long(evir_ASCII[k])):
                            num = k
                            break
                if(num != None):
                    tmp.append(1)
                else:
                    tmp.append(0)
            df.loc[i,'environment'] = str(tmp)
    return df
```

```python
def Utili2HotKey(df):
    for i in range(len(df)):
        if(pd.isnull(df.loc[i]['utility'])):
            continue
        else:
            tmp = []
            utility_list = df.loc[i]['utility'].replace(' ,', "").split(',')
            for k in range(len(utility_ASCII)):
                num = None
                for utility in utility_list:
                    if(utility != ''):
                        if(long(ASCII(utility)) == long(utility_ASCII[k])):
                            num = k
                            break
                if(num != None):
                    tmp.append(1)
                else:
                    tmp.append(0)
            df.loc[i,'utility'] = str(tmp)
    return df
```

```python
def compute_mean_mode(exam_df, input_df, index):
    tmp = exam_df[exam_df.surface <= (input_df.loc[index]['surface'] + input_df.loc[index]['surface']*0.1)]
    tmp = tmp[tmp.surface >= (input_df.loc[index]['surface'] - input_df.loc[index]['surface']*0.1)]
    tmp = tmp[tmp.district == input_df.loc[index]['district']]

    dic = {}
    for column in tmp:
        column_notnull = tmp[tmp[column].notnull()]
        if(len(column_notnull) == 0):
            num = None
        else:
            num = column_notnull.index.get_values()[0]
        
        if(num != None):
            if (isinstance(tmp.loc[num][column], (str, unicode))):
                try:
                    dic[column] = str(tmp[column].mode()[0])
                except:
                    dic[column] = tmp.loc[num][column]
            else:
                dic[column] = tmp[column].mean()
        else:
            dic[column] = None
    return dic
```

```python
def zero2NaN(df):
    tmp = df[df.notnull().all(axis=1)]
    tmp1 = df[df.floor == 0]
    tmp2 = df[df.bedRoom == 0]
    tmp3 = df[df.bathRoom == 0]
    index = tmp1.index.get_values()
    for i in index:
        df.loc[i,'floor'] = np.NAN

    index = tmp2.index.get_values()
    for i in index:
        df.loc[i,'bedRoom'] = np.NAN

    index = tmp3.index.get_values()
    for i in index:
        df.loc[i,'bathRoom'] = np.NAN
    return df
```

```python
def fill_NaN(train):
    train_df = read_CSV('train_v1_1.csv')
    for column in train:
        df = train[pd.isnull(train[column])]
        index = df.index.get_values()
        print str(column) + ', len: ' + str(len(index))
        for i in index:
            if(pd.isnull(train.loc[i, column])):
                mean_mode_dict = compute_mean_mode(train_df, train, i)
                if(mean_mode_dict[column] == None):
                    key_name = ['kind','Legal','floor','bedRoom','bathRoom','environment','utility']
                    if(column == key_name[0]):
                        tmp = train_df[column].mode()[0]
                        train.loc[i, column] = tmp
                    if(column == key_name[1]):
                        tmp = train_df[column].mode()[0]
                        train.loc[i, column] = tmp
                    if(column == key_name[2]):
                        train.loc[i, column] = train_df[column].mean()
                    if(column == key_name[3]):
                        train.loc[i, column] = train_df[column].mean()
                    if(column == key_name[4]):
                        train.loc[i, column] = train_df[column].mean()
                    if(column == key_name[5]):
                        tmp = train_df[column].mode()[0]
                        train.loc[i, column] = tmp
                    if(column == key_name[6]):
                        tmp = train_df[column].mode()[0]
                        train.loc[i, column] = tmp
                else:
                    train.loc[i, column] = mean_mode_dict[column]
    return train
```

```python
def split_coordinates(df):
    X = []
    Y = []
    for i in range(len(df)):
        a = (df.loc[i]['coordinates'].replace('[', "").replace(']', "").split(', '))
        X.append(float(a[0]))
        Y.append(float(a[1]))
        
    df.insert(22, 'coordinatesX', X, allow_duplicates = True)
    df.insert(23, 'coordinatesY', Y, allow_duplicates = True)
    df.drop('coordinates', axis=1, inplace=True)
    return df
```

```python
def split_kind(df):
    townhouse = []
    other = []
    villa = []
    
    tmp = []
    
    for i in range(len(df)):
        a = (df.loc[i]['kind'].replace('[', "").replace(']', "").split(', '))
        townhouse.append(float(a[0]))
        other.append(float(a[1]))
        villa.append(float(a[2]))
       
    tmp.append(townhouse)
    tmp.append(other)
    tmp.append(villa)
    
    num = 24
    for i in range(len(kind_namekey)):
        df.insert(num, kind_namekey[i], tmp[i], allow_duplicates = True)
        num += 1
    df.drop('kind', axis=1, inplace=True)
    return df
```

```python
def split_legal(df):
    CHO = []
    CLUR = []
    legal_namekey = ['CHO', 'CLUR']
    
    tmp = []
    
    for i in range(len(df)):
        a = (df.loc[i]['Legal'].replace('[', "").replace(']', "").split(', '))
        CHO.append(float(a[0]))
        CLUR.append(float(a[1]))
       
    tmp.append(CHO)
    tmp.append(CLUR)
    
    num = 27
    for i in range(len(legal_namekey)):
        df.insert(num, legal_namekey[i], tmp[i], allow_duplicates = True)
        num += 1
    df.drop('Legal', axis=1, inplace=True)
    return df
```

```python
def split_Evir(df):
    scholl = []
    supermarket = []
    hospital = []
    publicTraffic = []
    temple = []
    church = []
    beach = []
    park = []
    security = []
    
    tmp = []
    
    for i in range(len(df)):
        a = (df.loc[i]['environment'].replace('[', "").replace(']', "").split(', '))
        scholl.append(float(a[0]))
        supermarket.append(float(a[1]))
        hospital.append(float(a[2]))
        publicTraffic.append(float(a[3]))
        temple.append(float(a[4]))
        church.append(float(a[5]))
        beach.append(float(a[6]))
        park.append(float(a[7]))
        security.append(float(a[8]))
       
    tmp.append(scholl)
    tmp.append(supermarket)
    tmp.append(hospital)
    tmp.append(publicTraffic)
    tmp.append(temple)
    tmp.append(church)
    tmp.append(beach)
    tmp.append(park)
    tmp.append(security)
    
    num = 33
    for i in range(len(evir_namekey)):
        df.insert(num, evir_namekey[i], tmp[i], allow_duplicates = True)
        num += 1
    df.drop('environment', axis=1, inplace=True)
    return df
```

```python
def split_utility(df):
    airCondition = []
    pool = []
    washingMachine = []
    internet = []
    terrace = []
    furniture = []
    wifi = []
    garden = []
    gym = []
    
    tmp = []
    
    for i in range(len(df)):
        a = (df.loc[i]['utility'].replace('[', "").replace(']', "").split(', '))
        airCondition.append(float(a[0]))
        pool.append(float(a[1]))
        washingMachine.append(float(a[2]))
        internet.append(float(a[3]))
        terrace.append(float(a[4]))
        furniture.append(float(a[5]))
        wifi.append(float(a[6]))
        garden.append(float(a[7]))
        gym.append(float(a[8]))
       
    tmp.append(airCondition)
    tmp.append(pool)
    tmp.append(washingMachine)
    tmp.append(internet)
    tmp.append(terrace)
    tmp.append(furniture)
    tmp.append(wifi)
    tmp.append(garden)
    tmp.append(gym)
    
    num = 42
    for i in range(len(utility_namekey)):
        df.insert(num, utility_namekey[i], tmp[i], allow_duplicates = True)
        num += 1
    df.drop('utility', axis=1, inplace=True)
    return df
```
